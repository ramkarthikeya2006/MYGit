```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read uploaded image
img = cv2.imread(list(uploaded.keys())[0])

# 1. Preprocessing (Grayscale + Smoothing)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smooth = cv2.GaussianBlur(gray, (5,5), 0)

# 2. HOG + Sliding Window (Pedestrian Detection)
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

boxes, _ = hog.detectMultiScale(smooth, winStride=(4,4), padding=(8,8), scale=1.05)

# 3. Visualization (Bounding Boxes)
for (x,y,w,h) in boxes:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)

# Show result in Colab
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
# ==============================
# STEP 1: INSTALL LIBRARIES
# ==============================
!pip install ultralytics opencv-python matplotlib

# ==============================
# STEP 2: IMPORT LIBRARIES
# ==============================
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# STEP 3: LOAD MODEL
# ==============================
model = YOLO("yolov8n.pt")  # lightweight pretrained model

# ==============================
# STEP 4: UPLOAD IMAGE
# ==============================
from google.colab import files
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# ==============================
# STEP 5: DETECTION (PERSON CLASS = 0)
# ==============================
results = model(image)

pred_boxes = []

for r in results:
    for box in r.boxes:
        cls = int(box.cls[0])

        if cls == 0:  # person class
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            pred_boxes.append([x1, y1, x2, y2])

            # Draw bounding box
            cv2.rectangle(image, (x1,y1), (x2,y2), (0,255,0), 2)

# ==============================
# STEP 6: SHOW OUTPUT IMAGE
# ==============================
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Pedestrian Detection")
plt.axis("off")
plt.show()

# ==============================
# STEP 7: IoU FUNCTION
# ==============================
def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter_area = max(0, x2-x1) * max(0, y2-y1)

    box1_area = (box1[2]-box1[0]) * (box1[3]-box1[1])
    box2_area = (box2[2]-box2[0]) * (box2[3]-box2[1])

    union_area = box1_area + box2_area - inter_area

    if union_area == 0:
        return 0
    return inter_area / union_area

# ==============================
# STEP 8: DEFINE GROUND TRUTH
# (EDIT THESE VALUES BASED ON IMAGE)
# ==============================
ground_truth = [
    [50, 60, 200, 300],   # Example box 1
    [300, 80, 450, 350]   # Example box 2
]

# ==============================
# STEP 9: CALCULATE IoU
# ==============================
print("\nIoU Values:")
for gt in ground_truth:
    for pred in pred_boxes:
        iou = compute_iou(gt, pred)
        print("GT:", gt, "Pred:", pred, "IoU:", round(iou, 3))

# ==============================
# STEP 10: CALCULATE RECALL
# ==============================
TP = 0

for gt in ground_truth:
    matched = False
    for pred in pred_boxes:
        if compute_iou(gt, pred) > 0.5:
            matched = True
            break
    if matched:
        TP += 1

recall = TP / len(ground_truth) if len(ground_truth) > 0 else 0

print("\nRecall:", round(recall, 3))

# ==============================
# STEP 11: ANALYSIS OUTPUT
# ==============================
print("\n--- Analysis ---")
print("Crowded Scenes: Overlapping people reduce accuracy.")
print("Occlusions: Hidden persons may not be detected → recall decreases.")
print("YOLO is fast but struggles with dense crowds.")
```
